```python
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

def log_ood(income):
  return(-12+0.25*income)

income=40
z=log_ood(income)
p=sigmoid(z)
print("Income:",income)
print("z=",z)
print("probability of loan approval",p)

target_p=0.90
z_target=math.log(target_p/(1-target_p))
print("z=",z_target)
income_need=(z_target+12)/0.25
print("Income needed:",income_need)
print("Income in thousands",income_need*1000)
```
